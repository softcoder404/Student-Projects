### Feature Engineering

- Date Time features
- Lag features
- Window features
- Expanding features

### Date Time features

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/mac/Documents/dataframe1.csv', header=0, parse_dates=[0])
```

```python
df.head()
```

```python
feature = df.copy()
```

```python
feature['year'] = df['DATE'].dt.year
```

```python
feature['month'] = df['DATE'].dt.month
```

```python
feature['day'] = df['DATE'].dt.day
```

```python
feature.head()
```

### Lags features

To create lag1 feature

```python
feature['lag1'] = df['H1'].shift(1)
```

```python
feature['lag2'] = df['H1'].shift(7)
```

```python
feature.head(10)
```

### Window features

```python
feature['Roll_mean'] = df['H1'].rolling(window = 2).mean()
```

```python
feature.head()
```

```python
feature['Roll_max'] = df['H1'].rolling(window = 3).max()
```

```python
feature.head(10)
```

## Expanding features

```python
feature['Expand_max'] = df['H1'].expanding().max()
```

```python
feature.head(10)
```
